```python
import torch
from torch.nn.parameter import Parameter
```

```python
input = torch.tensor([[[[1,2,0,1],
                        [-1,0,3,2],
                        [1,3,0,1],
                        [2,-2,1,0]]]]).float()
print(input)
print(input.shape)
```

```python
conv2d = torch.nn.Conv2d(in_channels=1,
                         out_channels=2,
                         kernel_size=2,
                         stride=1,
                         bias=False)
conv2d.weight = Parameter(torch.tensor([[[[1,-1],[-1,1]]]]).float())
print(conv2d)
print(conv2d.weight)
```

```python
output1 = conv2d(input)
print(output1)
```

```python
conv2d = torch.nn.Conv2d(in_channels=1,
                         out_channels=2,
                         kernel_size=2,
                         stride=2,
                         bias=False)
conv2d.weight = Parameter(torch.tensor([[[[1,-1],[-1,1]]]]).float())
output2 = conv2d(input)
print(output2)
```

```python
input = torch.tensor([[[[1,2,0,1],
                        [5,0,3,2],
                        [4,3,0,1],
                        [8,-2,1,0]]]]).float()
```

```python
conv2d = torch.nn.Conv2d(in_channels=1,
                         out_channels=2,
                         kernel_size=3,
                         stride=1,
                         padding=1,
                         bias=False)
conv2d.weight = Parameter(torch.tensor([[[[2,1,-1],[1,-1,1],[3,1,2]]]]).float())
output2 = conv2d(input)
print(output2)
```

```python
maxpool = torch.nn.MaxPool2d(2)
output = maxpool(output2)
print(output)
```

```python
import matplotlib.pyplot as plt
x = torch.linspace(-10,10,100)
relu = torch.nn.ReLU()
y = relu(x)
plt.plot(x.numpy(), y.numpy())
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.title('ReLU Function')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = torch.linspace(-10,10,100)
sigmoid = torch.nn.Sigmoid()
y = sigmoid(x)
plt.plot(x.numpy(), y.numpy())
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.title('ReLU Function')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = torch.linspace(-10,10,100)
softmax = torch.nn.Softmax()
y = softmax(x)
plt.plot(x.numpy(), y.numpy())
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.title('ReLU Function')
plt.grid(True)
plt.show()
```
